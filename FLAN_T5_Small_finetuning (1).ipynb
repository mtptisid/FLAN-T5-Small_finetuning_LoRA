```python
# Set up quantization for memory efficiency
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=False
)
```

```python
import torch
import bitsandbytes
import triton
print("Torch Version:", torch.__version__)  # Should be 2.3.1
print("BitsAndBytes Version:", bitsandbytes.__version__)  # Should be 0.43.3
print("Triton Version:", triton.__version__)  # Should be 2.3.1
print("CUDA Available:", torch.cuda.is_available())  # Should be True
print("GPU Name:", torch.cuda.get_device_name(0))  # Should be "NVIDIA T4"
```

```python
# Load tokenizer and model
model_name = "google/flan-t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16
)
```

```python
# Verify GPU availability
print("CUDA Available:", torch.cuda.is_available())
print("GPU Name:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```

# Apply LoRA for parameter-efficient fine-tuning

```python
# Apply LoRA for parameter-efficient fine-tuning
lora_config = LoraConfig(
    r=8,  # LoRA rank
    lora_alpha=32,
    target_modules=["q", "v"],  # Target attention modules
    lora_dropout=0.1,
    bias="none",
    task_type="SEQ_2_SEQ_LM"
)
model = get_peft_model(model, lora_config)
```

# Load dataset from Hugging Face

```python
from datasets import load_dataset, disable_caching
try:
    dataset = load_dataset("mtpti5iD/redhat-docs_dataset", cache_dir=None)
    print("Dataset loaded successfully:", dataset)
except Exception as e:
    print("Error loading dataset:", e)
```

```python


# Handle null values and format data
def format_example(example):
    command = example['commands'] if example['commands'] is not None else ""
    title = example['title'] if example['title'] is not None else "Untitled"
    content = example['content'] if example['content'] is not None else ""
    return {
        "text": f"Title: {title}\nContent: {content}\nCommand: {command}".strip()
    }

dataset = dataset.map(format_example)
```

# Tokenize dataset

```python
# Tokenize dataset
def tokenize_function(examples):
    inputs = tokenizer(
        examples['text'],
        max_length=512,
        truncation=True,
        padding="max_length"
    )
    inputs["labels"] = inputs["input_ids"].copy()
    return inputs

tokenized_dataset = dataset.map(tokenize_function, batched=True)
tokenized_dataset = tokenized_dataset.remove_columns(["text", "title", "content", "commands", "url"])
```

```python
# Set up data collator
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    fp16=True,
    optim="paged_adamw_8bit",
    logging_steps=100,
    save_steps=500,
    save_total_limit=2,
    report_to="none",
)
```

#  Initialize and train

```python
#  Initialize and train
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    data_collator=data_collator
)

trainer.train()

```

```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration
from peft import PeftModel
from datasets import load_dataset
from huggingface_hub import login
import json
import os

# step1
if hf_token:
    login(token=hf_token)
    print("Logged in to Hugging Face.")

# Step 2: Verify GPU
print("Torch Version:", torch.__version__)
print(f"GPU: {torch.cuda.get_device_name() if torch.cuda.is_available() else 'No GPU'}")

# Step 3: Load model and tokenizer
hf_repo_id = "mtpti5iD/flan-t5-small-redhat-docs-v2"  # Replace with your repo
base_model_name = "google/flan-t5-small"
try:
    tokenizer = T5Tokenizer.from_pretrained(hf_repo_id, subfolder="model", token=hf_token)
    base_model = T5ForConditionalGeneration.from_pretrained(
        base_model_name,
        device_map="cuda",
        torch_dtype=torch.bfloat16,
        load_in_4bit=True
    )
    model = PeftModel.from_pretrained(base_model, hf_repo_id, subfolder="model", token=hf_token)
    print("Loaded LoRA model and tokenizer from Hugging Face.")
except Exception as e:
    raise Exception(f"Failed to load model from {hf_repo_id}: {e}")

# Step 4: Inference function
def run_inference(question, context, max_length=100):
    prompt = f"Question: {question}\nContext: {context}"
    inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True).to("cuda")
    outputs = model.generate(**inputs, max_length=max_length)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Step 5: Custom question
custom_question = input("Enter your question (e.g., 'How do I install Package X?'): ")
custom_context = input("Enter context (e.g., 'Title: Installing Package X\nContent: To install Package X, use the package manager yum.\nCommand: yum install X'): ")
custom_output = run_inference(custom_question, custom_context)
print(f"\nCustom Question: {custom_question}")
print(f"Context: {custom_context}")
print(f"Output: {custom_output}")

# Step 6: Predefined test cases
test_cases = [
    {
        "question": "How do I install Package xlibs?",
        "context": "Title: Installing Package X\nContent: To install Package X, use the package manager yum. Ensure you have the correct repository configured.\nCommand: yum install X",
        "expected": "Run `yum install X`"
    },
    {
        "question": "How do I configure Service ntpd?",
        "context": "Title: Configuring Service Y\nContent: Edit the configuration file at /etc/y.conf and restart the service.\nCommand: systemctl restart y",
        "expected": "Edit /etc/y.conf and run `systemctl restart y`"
    },
    {
        "question": "What is the command to update the system?",
        "context": "Title: System Updates\nContent: Use the yum package manager to update all packages.\nCommand: yum update",
        "expected": "Run `yum update`"
    },
    {
        "question": "How do I start Service chronyd?",
        "context": "Title: Starting Service Z\nContent: Use systemctl to start the service.\nCommand: systemctl start z",
        "expected": "Run `systemctl start z`"
    },
    {
        "question": "How do I check Package W status?",
        "context": "Title: Checking Package W Status\nContent: Use rpm to verify if Package W is installed.\nCommand: rpm -q w",
        "expected": "Run `rpm -q w`"
    }
]

# Step 7: Dataset-derived test cases
jsonl_path = "redhat-docs_dataset.jsonl"  # Replace with your JSONL path
dataset_tests = []
if os.path.exists(jsonl_path):
    dataset = load_dataset('json', data_files=jsonl_path)
    print("Dataset loaded from JSONL.")

    def format_example(example):
        command = example['command'] if example['command'] is not None else ""
        title = example.get('title', 'Untitled')
        content = example.get('content', '')
        return f"Title: {title}\nContent: {content}\nCommand: {command}".strip()

    test_count = 0
    for example in dataset["train"]:
        if example.get("command") and test_count < 5:
            context = format_example(example)
            question = f"What is the command for {example.get('title', 'this task')}?"
            dataset_tests.append({
                "question": question,
                "context": context,
                "expected": example["command"]
            })
            test_count += 1
        if test_count >= 5:
            break
else:
    print("JSONL file not found. Trying Hugging Face dataset...")
    try:
        dataset = load_dataset(f"{hf_repo_id}", data_files="data/redhat-docs_dataset.jsonl")
        # Same dataset processing
    except Exception as e:
        print(f"No dataset available: {e}. Using predefined tests.")

# Step 8: Run tests
all_tests = test_cases + dataset_tests
results = []
for i, test in enumerate(all_tests):
    output = run_inference(test["question"], test["context"])
    pass_status = output.strip().lower() == test["expected"].strip().lower()
    results.append({
        "Test Case": i + 1,
        "Question": test["question"],
        "Context": test["context"],
        "Output": output,
        "Expected": test["expected"],
        "Pass": pass_status
    })
    print(f"\nTest Case {i + 1}:")
    print(f"Question: {test['question']}")
    print(f"Context: {test['context']}")
    print(f"Output: {output}")
    print(f"Expected: {test['expected']}")
    print(f"Pass: {pass_status}")

# Step 9: Calculate accuracy
accuracy = sum(1 for r in results if r["Pass"]) / len(results) * 100
print(f"\nAccuracy: {accuracy:.2f}% ({sum(1 for r in results if r['Pass'])}/{len(results)} tests passed)")

# Step 10: Save results
results.append({"Custom Question": custom_question, "Custom Context": custom_context, "Custom Output": custom_output})
with open("accuracy_results.json", "w") as f:
    json.dump(results, f, indent=2)
print("Results saved to accuracy_results.json")
```
